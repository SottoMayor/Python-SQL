```python
# To import the data we can use the pyodbc as made in the main file...
# But, I chose a different approach. Let's import the salaries.sqlite via...
# sqlite3 package and read the data with pandas.

import sqlite3
import pandas as pd

connection = sqlite3.connect('salarios.sqlite') # Opening the connection
dataframe = pd.read_sql('SELECT * FROM salaries', connection)
connection.close() # Closing the connection

# display(dataframe) # Great! We are good to go.
```

```python
# What has been the evolution of the average wage over the years?
df_average_years = dataframe.groupby(by='Year').mean()[['TotalPay', 'TotalPayBenefits']]

display(df_average_years)

df_average_years.plot()

```
